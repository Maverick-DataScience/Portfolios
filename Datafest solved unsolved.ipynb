Here is few conditions to decide if the case is considered as unsolved, solved, or not taken:

Unsolved if:
- if the case is taken by an attorney or volunteer, but the legal deadline date is before the closed question date
- if the case is taken by an attorney or volunteer, but there's no further update

Not taken if:
- if the case is not taken 

Solved if:
- if the case is taken by an attorney or volunteer, and the point that is mentioned in the unsolved case doesn't happen
